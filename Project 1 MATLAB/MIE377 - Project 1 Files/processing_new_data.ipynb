```python
import os
import random
import pandas as pd

from datetime import datetime
```

# Find appropriate dates

```python
df = pd.read_csv('MIE377_AssetPrices.csv')

n = 1
dates_list = df['Date'].tolist()
#column = ['Adjusted Close']

formatted_dates = []

files = os.listdir('S&P')

assets = []

#Dates in AssetPrices dataframe
for date_str in dates_list:
    # Parse the date string
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    # Format the date object as "YYYY-MM-DD" and append it to the formatted dates list
    formatted_date = date_obj.strftime("%d-%m-%Y")
    formatted_dates.append(formatted_date)

print(formatted_dates)
#filtered_data.append(formatted_dates)

#get dates from A.csv - 127 dates in most S&P 500 stocks
filepath = os.path.join('S&P', 'A.csv')
df = pd.read_csv(filepath)
df = df[df['Date'].isin(formatted_dates)] #get correct dates
new_dates = df['Date'].tolist()

for i in files:
    filepath = os.path.join('S&P', i)
    df = pd.read_csv(filepath)
    df = df[df['Date'].isin(new_dates)] #get correct dates
    new_dates2 = df['Date'].tolist()

    # #new_dates = df.tolist()
    if new_dates2 == new_dates:
        #print(new_dates2)
        assets.append(files)

print(assets)
#print(len(assets))
```

# Choose Random Set of Assets from S&P 500

```python
filtered_data = []
n = 4

assets2 = []
random_files = random.sample(assets[0], 45)
print(random_files)

filtered_data.append(dates_list)

for file in random_files:
    filepath = os.path.join('S&P', file)
    df = pd.read_csv(filepath)
    df = df[df['Date'].isin(formatted_dates)] #get correct dates
    #prices = df['Adjusted Close'].values #get prices
    prices = df['Adjusted Close'].tolist()
    if (len(prices)) == 127:
        assets2.append(file)
        filtered_data.append(prices)


print(filtered_data)

for num in range (1,11):
    print(len(filtered_data[num]))
transposed_data = list(zip(*filtered_data))
# print(transposed_data)
# print(len(transposed_data))
data_new = pd.DataFrame(transposed_data)
data_new.to_csv('AssetPrices_{}'.format(n), index=False)
```
