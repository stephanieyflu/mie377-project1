```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
df_prices = pd.read_csv('MIE377_AssetPrices1.csv')
```

```python
df_prices.columns
```

```python
df_prices
```

```python
fig, ax = plt.subplots(7, 3, figsize=(12, 20))

for i in range(1, len(df_prices.columns)):
    j = (i - 1) // 3  # Calculate row index
    k = (i - 1) % 3   # Calculate column index
    ax[j, k].plot(df_prices['Date'].astype('datetime64[D]'), df_prices[df_prices.columns[i]])
    ax[j, k].set_title(df_prices.columns[i])
    ax[j, k].set_ylabel('Asset Price')

plt.tight_layout()
plt.show()
```
